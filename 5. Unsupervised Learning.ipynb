# 2. K-평균 알고리즘
앞의 경우에는 사과, 파인애플, 바나나 사진임을 미리 알고 있었기에 각 과일의 평균을 구할 수 있었다. 하지만 어떤 과일이 있는지 모르는 경우가 대부분이다.

이런 경우 **K-평균(K-means)** 군집 알고리즘이 자동적으로 평균값을 찾아준다. 이 평균값이 클러스터 중심에 위치하기 때문에 **클러스터 중심(Cluster Center)** 또는 **센트로이드(Centroid)** 라고 부른다.

 K-평균 알고리즘 작동 방식은 다음과 같다.
 1. 무작위로 k개의 클러스터 중심을 정한다.
 1. 각 샘플에서 가장 가까운 클러스터 중심을 찾아 클러스터의 샘플로 지정한다.
 1. 클러스터에 속한 샘플의 평균값으로 클러스터 중심을 변경한다.
 4. 클러스터 중심에 변화가 없을 때까지 2번으로 돌아가 반복한다.
 
 <img src='img/5/k-means.jpg'>

```python
fruits_2d = fruits.reshape(-1, 100*100)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3, random_state=42) 
# 클래스 객체는 labels_ 속성에 저장하며 배열 길이는 샘플 개수와 동일
# n_cluster=3 으로 샘플이 3가지이기 때문에 labels_ 배열의 값은 0, 1, 2 중 하나
km.fit(fruits_2d)
```

```python
print(km.labels_)
# 0, 1, 2 와 레이블 순서는 의미 X
```

```python
print(np.unique(km.labels_, return_counts=True))
```

```python
import matplotlib.pyplot as plt
def draw_fruits(arr, ratio=1):
    n = len(arr) # n은 샘플 개수
    # 한 줄에 10개씩 이미지를 그림. 샘플 개수를 10으로 나누어 전체 행 개수 계산
    rows = int(np.ceil(n/10)) # np.ceil : 올림
    # 행이 1개면 열의 개수는 샘플 개수. 그렇지 않으면 10개
    cols = n if rows < 2 else 10
    fig, axs = plt.subplots(rows, cols, figsize=(cols*ratio, rows*ratio), squeeze=False)
    for i in range(rows):
        for j in range(cols):
            if i*10 + j < n:
                axs[i, j].imshow(arr[i*10 + j], cmap='gray_r')
            axs[i, j].axis('off')
    plt.show()
```

```python
draw_fruits(fruits[km.labels_==0]) # 파인애플
# km.labels_==0 을 주면 레이블이 0인 값은 True, 그 외는 False 가 되어 특정 원소만 선택하는 것이 가능
# 이를 불리언 인덱싱이라 함
```

```python
draw_fruits(fruits[km.labels_==1])
```

```python
draw_fruits(fruits[km.labels_==2])
```

### 2-1. 클러스터 중심
KMeans 가 최종적으로 찾은 클러스터 중심은 cluster_centers_ 속성에 저장되어 있다.

이 배열은 fruits_2d 샘플의 클러스터 중심이기 때문에 각 중심을 이미지로 출력하려면 100 x 100 크기의 2차원 배열로 바꿔야 한다.

```python
draw_fruits(km.cluster_centers_.reshape(-1, 100, 100), ratio=3)
# 이전 절에서 사과, 바나나, 파인애플의 픽셀 평균값과 매우 유사
```

KMeans 클래스는 훈련 데이터 샘플에서 클러스터 중심까지 거리로 변환해주는 transform() 메서드를 가지고 있다. 그 말은 StandardScaler 클래스처럼 특성 값을 변환하는 도구로 사용할 수 있다는 의미이다.

fit() 메서드와 마찬가지로 2차원 배열을 기대한다.

```python
print(km.transform(fruits_2d[100:101]))
# fruits_2d[100] 은 (10000,) 형태로 전달되기에 에러가 발생
# 슬라이싱 연산자를 사용하여 (1, 10000) 형태로 전달

# 레이블 0, 1, 2 에 대한 각각의 거리 반환
# 여기서는 0이 제일 가까우므로 레이블 0에 속한다고 추측할 수 있음
```

```python
print(km.predict(fruits_2d[100:101]))
```

```python
print(km.n_iter_)
# K-Means 알고리즘은 반복적으로 클러스터 중심을 옮기면서 최적의 클러스터를 탐색
# 알고리즘이 반복한 횟수가 n_iter_ 속성에 저장됨
```

우리는 n_cluster=3 으로 지정하였다. 이것은 일종의 타깃에 대한 정보를 활용한 셈이다.

### 2-2. 최적의 K 찾기
 K-Means 알고리즘의 단점 중 하나는 클러스터 개수를 사전에 지정해야한다는 것이다. 실전에서는 몇 개의 클러스터가 있는지 알 수 없다.

 먼저 완벽하진 않지만 적절한 클서트러 개수를 찾기 위한 대표적인 방법 **엘보우(Elbow)** 방법이 있다.
 
 K-Means 알고리즘은 클러스터 중심과 클러스터에 속한 샘플 사이의 거리를 잴 수 있다. 이 거리의 제곱의 합을 **이너셔(Inertia)** 라고 부른다. 이너셔는 클러스터에 속한 샘플이 얼마나 가깝게 모여 있는지를 나타내는 값으로 생각할 수 있다. 클러스터의 개수가 늘어나면 클러스터 내 샘플의 개수도 줄어들기에 이너셔도 줄어든다.
 
 엘보우는 클러스터 개수를 늘려가면서 이너셔의 변화를 관찰하여 최적의 클러스터 개수를 찾는 방법이다.
 
 클러스터 개수를 증가시키면 이너셔를 그래프로 그렸을 때 감소하는 속도가 꺾이는 지점이 존재한다. 이 지점부터 클러스터 개수가 늘어나도 이너셔가 크게 줄어들지 않는다. 이것이 마치 팔꿈치 같다 하여 엘보우 방법이라 부른다.

```python
inertia = []
for k in range(2, 7):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(fruits_2d)
    inertia.append(km.inertia_) # inertia_ 속성에 저장된 이너셔 값을 inertia 리스트에 저장
plt.plot(range(2,7), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```

# 3. 주성분 분석

### 3-1. 차원 축소
데이터가 가진 속성을 특성이라 한다. 사진의 경우 10,000개의 픽셀이 있기에 10,000개의 특성이 있는 셈이다. 머신러닝에서 이런 특성을 **차원(Dimension)** 이라고도 부른다. 즉, 10,000개의 특성은 10,000개의 차원이라는 말인데 이 차원을 줄인다면 저장 공간을 절약할 수 있다.
>2차원 배열과 1차원 배열(벡터)에서 차원이란 용어는 다르게 사용된다. 다차원 배열에서 차원은 배열의 축 개수를 말한다. 가령 2차원 배열일 때는 행과 열이 차원이 된다. 하지만 1차원 배열, 즉 벡터의 경우 원소의 개수를 말한다.

![dimension](https://github.com/JuJab/Convolution-study/assets/98221614/ec2b6843-28ba-4f41-9178-eb9f10711cf9)

비지도 학습 작업 중 하나인 **차원 축소(Dimensionality Reduction)** 알고리즘이 있다. 차원 축소는 데이터를 가장 잘 나타내는 일부 특성을 선택하여 데이터 크기를 줄이고 지도 학습 모델의 성능을 향상시킬 수 있는 방법이다. 또한 줄어든 차원에서 다시 원본 차원으로 손실을 최대한 줄이면서 복원할 수도 있다. 대표적인 차원 축소 알고리즘인 **주성분 분석(Principal Component Analysis)** 가 있고, 간단히 **PCA** 라고도 부른다.

### 3-2. 주성분 분석 소개
PCA는 데이터의 분산이 큰 방향을 찾는 것으로 이해할 수 있다. 분산이 큰 방향이란 데이터를 잘 표현하는 어떤 벡터라고 생각할 수 있다. 데이터 x₁x₂2개의 특성이 있다. 이 때 분산이 가장 큰 방향, 즉 가장 데이터의 분포를 잘 표현하는 방향은 다음과 같다.

<img src='https://github.com/JuJab/Convolution-study/assets/98221614/640693a6-bbb8-40e5-b752-54c0f1ccdf8c' width='350' height='350'/>

직관적으로 그림과 같이 길게 늘어진 대각선 방향이 분산이 가장 크다고 볼 수 있다. 화살표의 위치와 방향은 큰 의미가 없다. 중요한 것은 분산이 큰 방향을 찾는 것이 중요하다.

<img src='https://github.com/JuJab/Convolution-study/assets/98221614/12dc1481-c389-4d9a-b36e-a3936ac2303e' width='350' height='350'/>

이러한 데이터가 있을 때 위치를 옮겨 원점에서 시작한다고 하면 (2, 1)과 같이 벡터를 표현할 수 있게 된다. 이 벡터를 **주성분(Principal Component)** 라고 한다. 주성분 데이터는 원본 데이터에 있는 어떤 방향이며, 원소 개수는 원본 데이터셋에 있는 특성 개수와 같다. 그렇기에 특성을 **차원** 이라고도 부르는 것이다.

> 사이킷런의 PCA 모델을 훈련하면 자동으로 특성마다 평균값을 빼서 원점에 맞추어 준다. 그렇기에 우리가 수동으로 데이터를 원점에 맞출 필요가 없다.

<img src='https://github.com/JuJab/Convolution-study/assets/98221614/f21397ee-5858-4036-b749-d9f9455af831' width='350' height='350'/>

원본 데이터는 주성분을 사용해 차원을 줄일 수 있다. 샘플 데이터 s(4, 2) 를 주성분에 직각으로 투영한다. 즉 샘플에서 주성분의 직선에 가장 가까운 점을 찾으면 1차원 데이터 p(4.5) 를 만들 수 있다.

주성분은 원본 차원과 같고 주성분으로 바꾼 데이터는 차원이 줄어든다. 주성분이 가장 분산이 큰 방향이기 때문에 주성분에 투영하여 바꾼 데이터는 원본이 가지고 있는 특성을 가장 잘 나타내고 있을 것이다.

<img src='https://github.com/JuJab/Convolution-study/assets/98221614/4b7f13c3-094c-4513-9cbb-057da09ab239' width='350' height='350'/>

첫 번째 주성분에 수직이며, 다음으로 분산이 가장 큰 방향이 두 번째 주성분이다. 수직이어야 주성분 서로의 방향에서 데이터 간섭이 완전히 없기 때문이다. 만약 수직이 아니면 방향을 수직으로 나누었을 때 첫 번째 주성분과 방향이 겹치는 데이터 벡터가 나올 것이다.

>기술적인 이유로 주성분은 원본 차원(특성)의 개수와 샘플 개수 중 작은 값만큼 찾을 수 있다. 일반적으로 비지도 학습은 대량의 데이터에서 수행하기에 원본 차원만큼 찾을 수 있다고 말한다.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=50) # n_components 는 주성분 개수
pca.fit(fruits_2d)
```

```python
print(pca.components_.shape) # components_ 속성에 주성분이 저장
```

```python
draw_fruits(pca.components_.reshape(-1, 100, 100))
```

```python
print(fruits_2d.shape)
```

```python
fruits_pca = pca.transform(fruits_2d)
print(fruits_pca.shape)
```

```python
fruits_inverse = pca.inverse_transform(fruits_pca)
# 원본 데이터로 복원은 가능. 하지만 어느 정도의 손실은 발생할 수밖에 없음
print(fruits_inverse.shape)
```

```python
fruits_reconstruct = fruits_inverse.reshape(-1, 100, 100)
for start in [0, 100, 200]:
    draw_fruits(fruits_reconstruct[start:start+100])
    print("\n")
```

### 3-3. 설명된 분산 (Explained Variance)
주성분이 원본 데이터의 분산을 얼마나 잘 나타내는지 기록한 값을 **설명된 분산(Explained variance)** 이라고 한다.

```python
print(np.sum(pca.explained_variance_ratio_))
# explained_variance_ratio_ 속성에 각 주성분의 설명된 분산 비율이 기록되어 있음
# 당연히 첫 번째 주성분의 설명된 분산이 가장 큼.
```

92% 가 넘는 분산을 유지하고 있다. 앞에서 50개의 특성에서 원본 데이터 복원 시 이미지 품질이 높았던 이유를 여기서 찾을 수 있다.

```python
plt.plot(pca.explained_variance_ratio_)
plt.show()
# 그래프를 보아 10개의 주성분으로도 충분히 높은 분산을 가짐을 확인할 수 있음
```
