```python
import pandas as pd

wine = pd.read_csv('https://bit.ly/wine_csv_data')

# https://www.kaggle.com/uciml/red-wine-quality-cortez-et-al-2009 일부 발췌
```

```python
wine.head()
# class 는 타깃값으로 0이면 레드와인, 1이면 화이트와인
```

```python
wine.info() # 6,497개의 샘플이 있고, 4개의 열이 모두 실수값인 것을 확인 가능
# 누락된 값 존재 시, 해당 데이터를 버리거나 평균값으로 채운 후 사용
```

```python
wine.describe()
# mean : 평균
# std : 표준편차
```

```python
data = wine[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)
# test_size 의 default 는 0.25(25%) 이다.
```

```python
print(train_input.shape, test_input.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_scaled, train_target)
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
print(lr.coef_, lr.intercept_)
```

# 1. 결정 트리 (Decision Tree)

 데이터를 잘 나눌 수 있는 질문을 탐색, 추가하여 분류 정확도 ↑
 
 <img src='img/4/decision_tree.jpg' width='250' height='250' align='left'/>

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
# 최적의 분할을 찾기 전에 특성의 순서 섞음
# 무작위성이 주입되고, 따라서 실행마다 점수가 조금씩 달라짐
dt.fit(train_scaled, train_target)
print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(10,7))
plot_tree(dt)
plt.show()
```

맨 위의 노드(node) 를 **루트 노드(root node)**

맨 아래의 노드(node) 를 **리프 노드(leaf node)**

> 노드란? 결정 트리를 구성하는 핵심 요소. 훈련 데이터의 특성에 대한 테스트를 표현하며, 가지(branch) 는 테스트의 결과(True, False) 를 나타내며 일반적으로 하나의 노드는 2개의 가지를 가진다.

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(10,7))
plot_tree(dt, max_depth=1, filled=True, feature_names=['alcohol','sugar','pH'])
# max_depth : 1을 주면 루트 노드를 제외하고 하나의 노드를 더 확장하여 그림
# filled : 클래스에 맞게 노드의 색을 칠할 수 있음. 한쪽 클래스의 비율이 높아질 수록 색이 진해짐
# feature_names : 특성의 이름 전달
plt.show()

# 바로 위의 노드를 부모 노드(parent node), 바로 아래의 노드를 자식 노드(child node) 라고 한다.
# 위의 Tree 에서는 sugar 를 기준으로 데이터를 분류하고 있다. 여러 특성 중 가장 데이터를 나누기 좋은 값을 사용.
# sample 은 데이터의 갯수, value 는 sample 데이터 중 음성 클래스(Red Wine) 와 양성 클래스(White Wine) 의 갯수를 표현해준다.
# gini 는 불순도 기준을 말한다.
```

최종적으로 리프 노드 중 가장 많은 클래스가 예측 클래스가 된다.

### 1-1. 지니 불순도(Gini Impurity)

gini 는 **지니 불순도(Gini Impurity)** 를 의미한다.

DecisionTreeClassifier 클래스의 criterion 매개변수의 default 값이 'gini'이며, 해당 매개변수는 노드에서 데이터를 분할할 기준을 정하는 것이다.

$지니 불순도 = 1 - (음성 클래스 비율^2 + 양성 클래스 비율^2)$

다중 클래스의 경우 클래스가 더 늘어나지만 계산방법은 동일하다. 돌아와서 우리의 예제로 다시 한 번 계산해보면,

$1 - ((1258 / 5197)^2 + (3939 / 5197)^2) = 0.367$

만약 노드의 클래스비율이이 정확히 반반인 경우 지니 불순도는 0.5가 되어 최악이 된다.

$1 - ((50 / 100))^2 + (50 / 100)^2) = 0.5$

반대로 노드에 하나의 클래스만 있다면 지니 불순도는 0이 되어 가장 작다. 이런 노드를 순수 노드라고 부른다.

$1 - ((0 / 100)^2 + (100 / 100)^2) = 0$

결정 트리 모델은 부모 노드(parent node) 와 자식 노드(child node) 의 불순도 차이가 가능한 크게 트리를 성장시킨다.

#### 부모 노드와 자식 노드의 불순도 차이 계산법

$부모의 불순도 - (왼쪽 노드 샘플 수 / 부모의 샘플 수) × 왼쪽 노드 불순도 - (오른쪽 노드 샘플 수 / 부모의 샘플 수) × 오른쪽 노드 불순도$

$0.367 - (2922 / 5197) × 0.481 - (2275 / 5197) × 0.069 = 0.066$

 이런 부모와 자식 노드 사이의 불순도 차이를 **정보 이득(Information Gain)** 이라고 한다. 그리고 해당 알고리즘은 정보 이득이 최대가 되도록 데이터를 나누고, 이 때 지니 불순도를 기준으로 한다.

 사이킷런에는 또 다른 불순도 기준이 존재한다.

 criterion='entropy' 를 지정하면 **엔트로피 불순도(Entropy Impurity)** 를 사용한다. 마찬가지로 클래스 비율을 사용하지만 제곱이 아닌 밑이 2인 로그를 사용하여 곱한다.
 
$음성 클래스 비율 × log₂(음성 클래스 비율) - 양성 클래스 비율 × log₂(양성 클래스 비율)$

$- (1258 / 5197) × log₂(1258 / 5197) - (3939 / 5197) × log₂(3939 / 5197) = 0.798$

지니 불순도와 엔트로피 불순도의 결과 차이는 크지 않다.

# 2. 교차 검증과 그리드 서치

### 2-1. 검증 세트 (Validation Set)

 테스트 세트를 자주 사용하면 결국 테스트 세트에 과대적합된 데이터가 될 확률이↑
 
 그렇기에 테스트 세트는 마지막에 한 번만 사용하는 것이 좋기에, 훈련 세트를 다시 나누어 임시 테스트 세트처럼 사용하는 방법이 있다. 이 데이터를 **검증 세트(Validation Set)** 이라 부른다.
 
<img src='img/4/validation_set.jpg' width='400' height='400'/>

>보통 20 ~ 30% 를 테스트 세트와 검증 세트로 떼어 놓으나, 훈련 데이터가 아주 많으면 단 몇 %만 떼어 놓아도 전체 데이터를 대표하는데 문제는 없다.

 훈련 세트로 모델을 훈련하고, 검증 세트로 모델을 평가한다. 이렇게 가장 좋은 모델을 고르면 해당 매개변수를 사용해 훈련 세트와 검증 세트를 합쳐, 전체 훈련 데이터에서 모델을 다시 훈련한다. 그리고 마지막에 테스트 세트에서 최종 점수를 평가한다.

```python
import pandas as pd
wine = pd.read_csv('https://bit.ly/wine_csv_data')
```

```python
data = wine[['alcohol','sugar','pH']].to_numpy()
target = wine['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
sub_input, val_input, sub_target, val_target = train_test_split(train_input, train_target, test_size=0.2, random_state=42)
```

```python
print(sub_input.shape, val_input.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(sub_input, sub_target)
print(dt.score(sub_input, sub_target))
print(dt.score(val_input, val_target))
# 해당 모델은 과대적합 되었음을 알 수 있음.
```

### 2-2. 교차 검증 (Cross Validation)
검증 세트로 인해 훈련 세트가 줄었다. 많은 데이터를 훈련에 사용할수록 좋은 모델이 만들어지지만, 검증 세트를 너무 조금 할당하면 검증 점수가 들쭉날쭉하고 불안정할 것이다.

 **교차 검증(Cross Validaiton)** 을 이용하면 안정적인 검증 점수와 훈련에 더 많은 데이터를 사용할 수 있다.
 
 교차 검증은 검증 세트를 떼어 평가하는 과정을 여러 번 반복하며, 이 점수를 평균을 내어 최종 검증 점수를 얻는다.
 
<img src='img/4/cross_validation.png' width='600' height='600'/>

>훈련 세트를 여러 부분으로 나눠 교차 검증을 수행하는 것을 **k-폴드 교차 검증(k-Fold Cross Validation)** 이라 한다. 위의 그림의 경우 다섯 부분으로 나누었으며 5-폴드 교차 검증이라 부를 수 있다. k-겹 교차 검증이라고도 부른다.

 보통 5-폴드 교차 검증이나 10-폴드 교차 검증을 많이 사용하며, 데이터의 80 ~ 90% 까지 훈련에 사용할 수 있다.

```python
from sklearn.model_selection import cross_validate
scores = cross_validate(dt, train_input, train_target) # 모델 객체, 훈련 세트 전달
print(scores)

# fit_time, score_time, test_score 키의 딕셔너리를 반환
# 앞의 2개는 각각 모델 훈련 시간과 검증 시간을 의미
# 5개의 숫자는 5-폴드 교차 검증임을 알려줌. 5-폴드 교차 검증이 default 값임.
# cv 매개변수에서 폴드 수를 변경 가능 : ex) cross_validate(dt, train_input, train_target, cv=10)
```

```python
import numpy as np
print(np.mean(scores['test_score'])) 
# 교차 검증의 최종 점수는 test_score
# 이름은 test_scroe 이지만 검증 폴드의 점수임
```

 cross_validate() 에서는 훈련 데이터를 섞지 않는다. 앞에서 train_test_split() 함수가 전체 데이터를 섞었기에 따로 섞는 과정이 필요하지 않다. 하지만 만약 교차 검증 시 훈련 세트를 섞으려면 **분할기(splitter)** 를 지정해야 한다.

 사이킷런의 분할기는 교차 검증 시 폴드를 어떻게 나눌지 결정해준다. cross_validate() 함수는 회귀 모델의 경우 KFold 분할기를 사용, 분류 모델의 경우 타깃 클래스를 골고루 나누기 위해 StratifiedKFold 분할기를 사용한다.

```python
from sklearn.model_selection import StratifiedKFold
scores = cross_validate(dt, train_input, train_target, cv=StratifiedKFold())
print(np.mean(scores['test_score']))
```

```python
# 만약 훈련 세트를 섞은 후 10-폴드 교차 검증 수행 시
splitter = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
scores = cross_validate(dt, train_input, train_target, cv=splitter)
print(np.mean(scores['test_score']))
```

### 2-3. 하이퍼파라미터 튜닝
머신러닝 모델에서 학습하는 파라미터를 모델 파라미터라고 부른다.

반면 모델이 학습할 수 없어서 사용자가 지정해야만 하는 파라미터를 하이퍼 파라미터라고 한다. 사이킷런과 같은 머신러닝 라이브러리 사용 시 이런 하이퍼파라미터는 모두 클래스나 매서드의 매개변수로 표현된다.

 하이퍼 파라미터의 튜닝은, 
 1. 라이브러리가 제공하는 기본값 그대로 사용해 모델을 훈련. 
 1. 검증 세트의 점수나 교차 검증을 통해 매개변수를 조금씩 수정. (모델마다 적게는 1 ~ 2개, 많게는 5 ~ 6개의 매개변수 제공) 
 1. 최종적으로 가장 좋은 점수의 모델을 선택.
>사람의 개입없이 하이퍼파라미터 튜닝을 자동으로 수행하는 기술을 'AutoML' 이라고 부른다.



매개변수 하나에서 최적값을 찾았다고 하여도, 다른 매개변수의 값을 변경하면 최적값은 변하게 된다. 그러므로 매개변수들을 모두 같이 바꿔가면서 최적값을 찾아야되지만, 매개변수가 많아질수록 이 과정은 복잡해진다.

for 문으로 구현하는 것도 가능하지만 사이킷런에서 제공하는 **그리드 서치(Grid Search)** 를 사용하여 최적값을 찾을 수 있다.

GridSearchCV 클래스는 하이퍼파라미터 탐색과 교차 검증을 한 번에 수행한다. 별도로 cross_validate() 함수를 호출할 필요가 없다.

```python
from sklearn.model_selection import GridSearchCV
params = {'min_impurity_decrease':[0.0001, 0.0002, 0.0003, 0.0004, 0.0005]} # 매개변수와 탐색할 값의 리스트
```

```python
gs = GridSearchCV(DecisionTreeClassifier(random_state=42), params, n_jobs=-1)
# cv 의 default 값 = 5
# 즉, 5-폴드 교차 검증을 수행함
# n_jobs 의 default 값 = 1
# 병렬 실행에 사용할 CPU 코어 수를 지정하며, -1은 시스템에 존재하는 모든 코어를 사용
```

```python
# cv가 5, 그리드 서치의 리스트 갯수가 5.
# 즉, 5×5 = 25 개의 모델을 훈련함
gs.fit(train_input, train_target)
```

```python
dt = gs.best_estimator_ # 모델들 중 검증 점수가 가장 높은 모델의 매개변수 조합으로 전체 훈련세트에서 다시 훈련한 모델
print(dt.score(train_input,train_target))
# 최적의 매개변수는 best_params_ 속성에 저장
```

```python
gs.best_params_
```

```python
# 각 매개변수에서 수행한 교차 검증의 평균 점수 → cv_results_ 속성의 'mean_test_score' 키에 저장
print(gs.cv_results_['mean_test_score']) # 첫번째 값이 가장 큼
```

1. 탐색한 최상의 매개변수에서 전체 훈련 세트를 사용해 최종 모델을 훈련한다. 이 모델 또한 그리드 서치 객체에 저장된다.

### 2-4. 랜덤 서치(Random Search)

 위의 값에서 매개변수 간격을 0.0001 혹은 1로 설정했는데, 이에 대한 근거가 없다.
 
 이렇게 매개변수 값의 범위나 간격을 정하기 어려울 때, 또 너무 많은 매개변수 조건으로 그리드 서치 수행 시간이 오래 걸릴 때 **랜덤 서치(Random Search)** 를 사용한다.
 
 랜덤 서치는 매개변수 값의 목록을 전달하는 것이 아닌 매개변수를 샘플링할 수 있는 확률 분포 객체를 전달한다. 해당 객체는 싸이파이(scipy) 라이브러리에서 지원한다.
 >**싸이파이(Scipy)** 라이브러리는 적분, 보간, 선형대수, 확률 등을 포함한 수치 계산 전용 라이브러리이다.

```python
from scipy.stats import uniform, randint
# uniform과 randint 클래스는 모두 주어진 범위에서 고르게 값을 뽑음
# 이를 ' 균등 분포에서 샘플링한다' 라고 말함.
# randint 는 정수값을, uniform 은 실수값을 뽑음
```

```python
rgen = randint(0, 10)
rgen.rvs(10)
```

```python
np.unique(rgen.rvs(1000), return_counts=True)
```

```python
ugen = uniform(0, 1)
ugen.rvs(10)
# 난수 발생기와 유사.
# 샘플링 횟수는 시스템 자원이 허락하는 범위 내에서 최대한 큰 편이 좋음.
```

# 3. 트리의 앙상블

- 정형 데이터(Structured Data)

어떠한 구조로 된 데이터를 뜻하며 CSV, Database, Excel 등의 데이터 형태로 저장하기 쉽다.
- 비정형 데이터(Unstructured Data)

텍스트 데이터, 사진, 영상, 음악 등과 같이 정형화되지 않은 데이터들을 뜻한다.
> 텍스트와 사진을 데이터베이스에 저장할 수는 있다. 대표적으로 NoSQL 데이터베이스는 엑셀, CSV에 담기 어려운 텍스트나 JSON 데이터 저장에 용이하다.

 정형 데이터에 가장 뛰어난 성과를 내는 알고리즘이 **앙상블 학습(Ensemble Learning)** 이다. 앙상블이라는 사전적 의미처럼 여러 가지의 모델을 합쳐서 하나의 결과를 내는 방법이다. 이 알고리즘은 대부분 결정 트리를 기반으로 만들어졌다.
 
 비정형 데이터는 7장에서 배울 신경망 알고리즘이 가장 적합하다. 비정형 데이터는 규칙성을 찾기 어려워 정통적인 머신러닝으로 모델을 만들기 까다롭다.

### 3-1. 랜덤 포레스트(Random Forest)
앙상블 학습의 대표 주자이다.

랜덤 포레스트는 결정 트리를 랜덤하게 만들어, 결정 트리(나무)의 **숲** 을 만든다.그리고 각 결정 트리의 예측을 사용해 최종 예측을 만든다.

 데이터는 우리가 입력한 훈련 데이터에서 랜덤하게 샘플을 추출하여 훈련 데이터를 만드는데, 한 샘플이 중복되어 추출될 수 있다.

 예시로 1,000개의 샘플이 있는 가방에서 100개의 샘플을 뽑는다면 1개의 샘플을 뽑고, 뽑은 샘플을 다시 가방에 넣는다. 이렇게 100개를 가방에서 뽑으면 중복된 샘플이 존재할 수 있다. 이러한 샘플을 **부트스트랩 샘플(Bootstrap Sample)** 이라고 한다.기본적으로 부트스트랩 샘플은 훈련 세트와 크기를 같게 만든다. 즉, 1,000개의 샘플이 있다면 중복하여 1,000개의 샘플을 뽑는다.

또한 각 노드 분할 시 전체 특성 중 일부 특성을 무작위로 고른 후 최선의 분할을 찾는다. 분류 모델인 RandomForestClassifier 은 기본적으로 전체 특성 개수의 제곱근만큼 특성을 선택한다. 4개의 특성이 있다면 노드마다 2개를 랜덤하게 선택하여 사용한다. 다만 회귀 모델인 RandomForestRegressor 은 전체 특성을 사용한다.

> 분류일 때는 각 트리의 클래스별 확률을 평균하여 가장 높은 확률을 가진 클래스를 예측으로 선택하고, 회귀일 때는 그저 각 트리의 예측을 평균한다.

 랜덤 포레스트는 랜덤하게 선택한 샘플과 특성을 사용하기에 과대적합을 막아주고, 검증 세트와 테스트 세트에서 안정적인 성능을 얻을 수 있다.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

wine = pd.read_csv('https://bit.ly/wine_csv_data')
data = wine[['alcohol','sugar','pH']].to_numpy()
target = wine['class'].to_numpy()
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=-1)
# return_train_score=True 로 하면 검증 점수뿐 아니라 훈련 세트에 대한 점수도 같이 반환함. default 는 False
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# 랜덤 포레스트는 결정 트리의 앙상블이기에 DecisionTreeClassifier가 제공하는 중요한 매개변수를 모두 제공함
# criterion,max_depth, max_features, min_samples_split, min_impurity_decrease, min_samples_leaf 등
rf.fit(train_input,train_target)
print(rf.feature_importances_)
```

훈련 세트에서 중복을 허용하여 부트스트랩 샘플을 만들기에, 포함되지 않고 남는 샘플이 있다. 이런 샘플을 **OOB(Out of Bag) 샘플** 이라고 한다. 이 샘플들을 이용하여 마치 검증 세트와 같이 훈련한 결정 트리를 평가할 수 있다.

이 점수를 얻기 위해서는 oob_score=True 로 설정해줘야한다. 기본값은 False 이다. 그러면 각 결정 트리의 OOB 점수를 평균내어 출력한다.

```python
rf = RandomForestClassifier(oob_score=True, n_jobs=-1, random_state=42)
rf.fit(train_input, train_target)
print(rf.oob_score_)
```

### 3-2. 엑스트라 트리 (Extra Tree)
 랜덤 포레스트와 매우 유사한데, 차이점이라고 하면 부트스트랩 샘플을 사용하지 않는다는 점이다. 즉, 각 결정 트리를 만들 때 전체 훈련 세트를 사용한다. 대신 노드를 분할 시 가장 좋은 분할을 찾는 것이 아니라 무작위로 분할한다.

 참고로 엑스트라 트리가 사용하는 결정 트리가 바로 splitter='random' 인 결정 트리이다.

```python
from sklearn.ensemble import ExtraTreesClassifier
et = ExtraTreesClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(et, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
et.fit(train_input, train_target)
print(et.feature_importances_)
```

### 3-3. 그레디언트 부스팅 (Gradient Boosting)
깊이가 낮은 결정 트리를 사용하여 이전 트리의 오차를 보완하는 방법이다. GradientBoostingClassifier 은 기본적으로 깊이가 3인 결정 트리 100개를 사용한다. 깊이가 낮기에 과대적합에 강하고 높은 일반화 성능을 기대할 수 있다.

해당 방법에서는 경사 하강법을 사용하여 트리를 앙상블에 추가한다. 분류는 로지스틱 손실 함수를 사용하고, 회귀에서는 평균 제곱 오차 함수를 사용한다.

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
gb = GradientBoostingClassifier(n_estimators=500, learning_rate=0.2, random_state=42)
# n_estimators : 결정 트리 갯수
# learning_rate(학습률) : 경사하강법 때 산을 내려가는 속도. default 값은 0.1
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
gb.fit(train_input, train_target)
print(gb.feature_importances_) # 특성별 중요도
```

>subsample 매개변수는 트리 훈련 시 훈련 세트의 비율 을 정하는 값으로 기본값은 1.0이다. 만약 subsample이 1보다 작으면 훈련 세트의 일부만을 사용하며, 일부 샘플은 랜덤하게 선택한다. 그렇기에 마치 확률적 경사 하강법이나, 미니배치 경사 하강법과 유사하다.

일반적으로 그레디언트 부스팅이 랜덤 포레스트보다 조금 더 높은 성능을 얻을 수 있지만, 랜덤 포레스트는 트리를 병렬로 불러올 수 있는 반면 그레디언트 부스팅은 트리를 만들고 손실 함수를 바탕으로 다음 트리를 순서대로 추가하기 때문에 훈련 속도가 느리다. 즉, GradientBoostingClassifier 에는 n_jobs 매개변수가 없다. 회귀버전은 GradientBoostingRegressor 이다.

### 3-4. 히스토그램 기반 그레디언트 부스팅 (Histogram-based Gradient Boosting)
입력 특성을 255 + 1 개의 구간으로 나눈다. 노드 분할 시 최적의 분할을 매우 빠르게 찾을 수 있다. 누락된 값을 1에 할당한다.

트리의 개수를 지정하는데 n_estimators 대신 부스팅 반복 횟수를 지정하는 max_iter 을 사용한다.

```python
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier

hgb = HistGradientBoostingClassifier(random_state=42)
scores = cross_validate(hgb, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

### Permutation Importance
각 샘플의 특성을 섞으므로서 특성의 중요도를 판단하는 방법이다.

본래 샘플의 점수보다 같은 샘플 내에서 첫 번째 특성을 섞었을 때의 점수가 본래 점수의 70%이고, 두 번째 특성을 섞었을 때의 점수가 80%라면 첫 번째 특성의 중요도가 더 높다고 판단할 수 있다.

```python
from sklearn.inspection import permutation_importance

hgb.fit(train_input, train_target)
result = permutation_importance(hgb, train_input, train_target, n_repeats=10, random_state=42, n_jobs=-1)
# n_repeats 는 랜덤하게 섞을 횟수로 default 값은 5
print(result.importances_mean)
# 각 특성을 섞었을 때의 정확도 하락 [1번 특성 정확도 하락, 2번 특성 정확도 하락, 3번 특성 정확도 하락]
```

```python
result = permutation_importance(hgb, test_input, test_target, n_repeats=10, random_state=42, n_jobs=-1)
# 훈련 세트에 대한 특성 중요도만 출력할 수 있는 feature_importances_ 와 달리 테스트 세트에 대하여 출력 가능
print(result.importances_mean) # 중요도 importances, 평균 importances+mean, 표준 편차 importances_std
```

```python
hgb.score(test_input, test_target)
```

```python
# XGBoost 는 사이킷런 이외에 그레디언트 부스팅 알고리즘을 구현한 라이브러리
from xgboost import XGBClassifier

xgb = XGBClassifier(tree_method='hist', random_state=42)
scores = cross_validate(xgb, train_input, train_target, return_train_score=True)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

랜덤 포레스트 : 성능이 좋고 안정적. 부트스트랩 샘플을 만들고 전체 특성 중 일부를 랜덤하게 선택해 결정 트리를 생성

엑스트라 트리 : 랜덤 포레스트와 유사. 그러나 부트스트랩 샘플을 사용하지 않고 노드 분할 시 최선이 아니라 랜덤하게 분할. 그래서 훈련 속도가 빠르지만 보통 더 많은 트리가 필요.

그레디언트 부스팅 : 깊이가 얕은 트리를 연속적으로 추가하여 손실을 최소화. 성능이 뛰어나지만 병렬 훈련이 불가하여 훈련 속도가 느림. 그래도 학습률 매개변수를 조정하여 모델의 복잡도를 제어 가능. 학습률 매개변수가 크면 복잡하고 훈련 세트에 과대적합된 모델을 얻을 수 있음.

히스토그램 기반 그레디언트 부스팅 : 가장 뛰어난 앙상블 학습으로 평가받는 알고리즘. 훈련 데이터 256개 구간으로 변환하여 사용하기에 노드 분할 속도가 매우 빠름. 사이킷런 뿐 아니라 XGBoost, LightGBM 등의 라이브러리가 존재.
